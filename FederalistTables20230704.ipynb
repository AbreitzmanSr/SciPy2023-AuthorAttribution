```python
from nltk.tokenize import word_tokenize

kills = [',','.',"''",'',';','-',')','(',':']

hamiltonDicts = [] #list of dictionaries containing word freq for each of Hamilton's Federalist Papers
madisonDicts = []
jayDicts = []
disputedDicts = []
jointDicts = []

def getDocDict(str1):
#returns a dictonary containing frequencies of any word in string
#e.g. str1 = 'quick brown fox is quick.'
# returns {quick:2, brown:1, fox:1, is:1}
  x = {}
  words = word_tokenize(str1.lower().strip())
  for b in words:
        if b in x:
            x[b]+=1
        else:
            x[b]=1
  return(x)

for a in hamilton:
    hamiltonDicts.append(getDocDict(a[1][0]))
    
for a in madison:
    madisonDicts.append(getDocDict(a[1][0]))
                        
for a in jay:
    jayDicts.append(getDocDict(a[1][0]))
    
for a in joint:
    jointDicts.append(getDocDict(a[1][0]))
    
for a in disputed:
    disputedDicts.append(getDocDict(a[1][0]))
    
print(len(hamiltonDicts),len(madisonDicts),len(jayDicts),len(jointDicts),len(disputedDicts))
```

```python
def wordsCount(docDicts):
    wc = 0
    for a in docDicts:
        for x in a:
            wc+=a[x]
    return(wc)
```

```python
hamiltonWordsCount=wordsCount(hamiltonDicts)
madisonWordsCount=wordsCount(madisonDicts)
jayWordsCount=wordsCount(jayDicts)
jointWordsCount=wordsCount(jointDicts)
disputedWordsCount=wordsCount(disputedDicts)
print(hamiltonWordsCount,madisonWordsCount,jayWordsCount,disputedWordsCount,jointWordsCount)
```

```python
#return fraction of docs by author containing word, and word freq per 1000 author words
def wordScore(word,authorDicts,authWords):
    totalDocs = 0
    totalFreq = 0
    for a in authorDicts:
        if (word in a):
            totalDocs += 1
            totalFreq += a[word]
    return([totalDocs/float(len(authorDicts)),(1000*totalFreq)/float(authWords)])
```

```python
from tabulate import tabulate
def wordTable(word):
    jointW = wordScore(word,jointDicts,jointWordsCount)
    hamiltonW = wordScore(word,hamiltonDicts,hamiltonWordsCount)
    madisonW = wordScore(word,madisonDicts,madisonWordsCount)
    disputedW = wordScore(word,disputedDicts,disputedWordsCount)
    jayW = wordScore(word,jayDicts,jayWordsCount)
    print('Usage Rates for Word: "'+word+'"')
    data = [['Hamilton',int(1000*hamiltonW[0])/10.0,int(1000*hamiltonW[1])/1000.0],
             ['Madison',int(1000*madisonW[0])/10.0,int(1000*madisonW[1])/1000.0],
            ['Joint',int(1000*jointW[0])/10.0,int(1000*jointW[1])/1000.0],
            ['Disputed',int(1000*disputedW[0])/10.0,int(1000*disputedW[1])/1000.0],
           ['Jay',int(1000*jayW[0])/10.0,int(1000*jayW[1])/1000.0]]
    print (tabulate(data, headers=["Author", "% Docs Containing Word", "Usage Per 1000 Words"]))
```

```python
def getAuthorDict(whichAuth,whichNum):
    if (whichAuth=='h'):
        return(hamiltonDicts[whichNum])
    else:
        if (whichAuth=='m'):
            return(madisonDicts[whichNum])
        else:
            if (whichAuth=='d'):
                return(disputedDicts[whichNum])
            else:
                return({})
            
def fillUnknowns(madison5,hamilton5):
    unknown = []
    for i,a in enumerate(hamiltonDicts):
      if (not ((['h',i] in madison5) or (['h',i] in hamilton5))): 
        unknown.append(['h',i])
    for i,a in enumerate(madisonDicts):
      if (not ((['m',i] in madison5) or (['m',i] in hamilton5))): 
        unknown.append(['m',i])
    for i,a in enumerate(disputedDicts):
      if (not ((['d',i] in madison5) or (['d',i] in hamilton5))): 
        unknown.append(['d',i])    
    for i,a in enumerate(jointDicts):
      if (not ((['j',i] in madison5) or (['j',i] in hamilton5))): 
        unknown.append(['j',i])  
    return(unknown)

def wordRate(dict1,word):
    wc = 0
    match_count = 0
    for a in dict1:
        wc+=dict1[a]
        if (a==word):
            match_count = dict1[a]
    return((1000.0*match_count)/wc)

def FederalistDecision():
    madison5 = []
    hamilton5 = []
    
    unknown = fillUnknowns(madison5,hamilton5)

    for a in unknown:
        x = getAuthorDict(a[0],a[1])
        for w in x:
            if (w == 'while'):
              hamilton5.append([a[0],a[1]])
    
    unknown = fillUnknowns(madison5,hamilton5)
   
    for a in unknown:
        x = getAuthorDict(a[0],a[1])
        if (('whilst' in x) and (wordRate(x,'whilst')>=.25)):
              madison5.append([a[0],a[1]])
    
    unknown = fillUnknowns(madison5,hamilton5)
    
    for a in unknown:
        x = getAuthorDict(a[0],a[1])
        if (('upon' in x) and (wordRate(x,'upon')>=.9)):
            hamilton5.append([a[0],a[1]])
        else:
            madison5.append([a[0],a[1]])
 
    unknown = fillUnknowns(madison5,hamilton5)
    
    print(unknown,'\n madison:',madison5,'\n hamilton:',hamilton5)
    
```

```python
#return usage rate per 1000 words of a target word
#e.g. if target=='upon' appears 3 times in a 1500 
#word essay, we return a rate of 2 per 1000 words.
def rate_per_1000(docDict,target):
    if (target in docDict):
        wordCount=0
        for a in docDict:
            wordCount+=docDict[a]
        return(1000*docDict[target]/wordCount)
    else:
        return(0)
    
def federalist_decison_tree(docDict):
    if ('while' in docDict):
        return('hamilton')
    else:
        if (rate_per_1000(docDict,'whilst') >= .25):
                return('madison')
        if (rate_per_1000(docDict,'upon') >= .9):
                return('hamilton')
        else:
                return('madison')
            
            
right = wrong = 0
for a in hamiltonDicts:
    if (federalist_decison_tree(a)=='hamilton'):
        right+=1
    else:
        wrong+=1

for a in madisonDicts:
    if (federalist_decison_tree(a)=='madison'):
        right+=1
    else:
        wrong+=1
        
madisonCount = 0
for a in disputedDicts:
    if (federalist_decison_tree(a)=='madison'):
       madisonCount += 1
        
print('accuracy:'+str(right/(right+wrong))+
      ', % disputed attributed to Madison:'+
      str(100*madisonCount/12.0))
```
